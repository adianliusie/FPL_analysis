## Detailed printout to see exact statistics after a player scores a goal

```python
player_dict = player_helper.player_stat_dict
print("[Assists, Clean Sheets, Goals, Minutes, team against, points]")

for player_name, player_stats in player_dict.items():
    if sum(player_stats)[2] > 0:
        print(player_name)
        print("[A, C, G, M, O, P]")

        for i in range(len(player_stats)-1):
            if player_stats[i][2] == 1:
                print(player_stats[i+1])
```

## Statistics of players after scoring for players with atleast 10 goals

```python
import numpy as np

np.set_printoptions(formatter={'float': lambda x: "{0:0.3f}".format(x)})

for player_name, player_stats in player_dict.items():
    if sum(player_stats)[2] > 9:
        print(player_name, sum(player_stats)[2], "[A, C, G, M, O, P]")


        average = sum(player_stats)/sum([min(1, i[3]) for i in player_stats])
        print("Average: ", average)
    
        counter = np.array([0,0,0,0,0,0])
        games = 0
        for i in range(len(player_stats)-1):
            if player_stats[i][2] == 1:
                counter = counter + player_stats[i+1]
                games += 1
    
        print("After goal: ", counter/games, "\n")
```

## Statistics of players after getting more than 5 points in 2 consecutive games

```python
import numpy as np

np.set_printoptions(formatter={'float': lambda x: "{0:0.3f}".format(x)})

for player_name, player_stats in player_dict.items():
    if sum(player_stats)[3] > 2000:
        print(player_name, sum(player_stats)[2], "[A, C, G, M, O, P]")


        average = sum(player_stats)/sum([min(1, i[3]) for i in player_stats])
        print("Average: ", average)
    
        counter = np.array([0,0,0,0,0,0])
        games = 0
        for i in range(len(player_stats)-2):
            if player_stats[i][5] >5:
                if player_stats[i][5] >5:
                    counter = counter + player_stats[i+1]
                    games += 1
    
        print("After 2 good games: ", counter/games, "\n")
```
